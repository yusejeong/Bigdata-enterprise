```python
import nltk
import os
from konlpy.tag import Twitter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
t=Twitter()
```

```python
f=open('풀무원.txt', encoding ='utf-8')
```

```python
data=f.read()
```

```python
tn=t.nouns(data)
```

```python
tn
```

```python
stop_words = ['풀무원','명','머스','증세','검','주','진심','오늘']
tn= [each_word for each_word in tn if each_word not in stop_words]
```

```python
ko=nltk.Text(tn)
```

```python
ko.vocab().most_common(50)
```

```python
dat=ko.vocab().most_common(50)
```

```python
tmp=dict(dat)
wordcloud=WordCloud(font_path='NanumGothic.ttf',
                   relative_scaling = 0.2,
                    background_color='white',
                   ).generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
